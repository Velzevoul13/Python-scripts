```python
import requests
from bs4 import BeautifulSoup
import pandas

r=requests.get("https://www.ampereanalysis.com/")
c=r.content
soup=BeautifulSoup(c,"html.parser")
all=soup.find_all("div",{"class":"col-sm-12"})
all2=soup.find_all("div",{"class":"col-sm-6 col-md-4 item infinite-item"})
lst1=[]
lst2=[]
lst3=[]
dict={}
dict2={}

pad=all[1].find("span",{"class":"pad-10"}).text
title=pad.split()[-1]
dict["Date"]=title
dict["Name"]=all[1].find("b").text
dict["Title"]=all[1].find("u").text
lst1.append(dict)

pad2=all[2].find("span",{"class":"pad-10"}).text
title2=pad2.split()[-1]
dict2["Date"]=title2
dict2["Name"]=all[2].find("b").text
dict2["Title"]=all[2].find("u").text
lst2.append(dict2)

base_url="https://www.ampereanalysis.com/blog/page"
for page in range(0,25,1):
    print(base_url+str(page))
    r=requests.get(base_url+str(page))
    c=r.content
    soup=BeautifulSoup(c,"html.parser")
    all=soup.find_all("div",{"class":"col-sm-6 col-md-4 item infinite-item"})
    for item in all:
        dict3={}
        pad3=item.find("span",{"class":"pad-10"}).text
        title3=pad3.split()[-1]
        dict3["Date"]=title3
        dict3["Name"]=item.find("span",{"class":"pad-10"}).find("b").text
        dict3["Title"]=item.find("div",{"class":"col-md-12"}).find("u").text
        lst3.append(dict3)

l=lst1+lst2+lst3
df=pandas.DataFrame(l)
df.to_csv("Output.csv")
```
